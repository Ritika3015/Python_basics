**2) What is an interpreter in Python?**
- Interpreter is a program that reads and execute python code line by line.
- Unlike compiled level languages(C or Java) which uses compiler, Python uses interpreter that runs directly from the source code without first converting into machine level code.

**3) What are pre-defined keywords in Python?**

**4) Can keywords be used as variable names?**
- No, keywords cannot be used as variable names since they hold a speial meaning and has specific purpose in python.

**5) What is mutability in Python?**
- Objects / containers whose state or values can be changed after they are created are called mutable objects and the phenomenon is known as mutability. For example, lists in python show mutability i.e. if we change any values in the list, it is possible. So, lists are mutable. They help manage memory efficiently.

**6) Why are lists mutable, but tuples are immutable?**
- Lists are mutable because they can be modified i.e. we can add, remove or update items in a list whereas tuples are immutable i. we cannot perform any changes in a tuple once created.

**7) What is the difference between “==” and “is” operators in Python?**
- '==' operator known as equality operator, checks identity i.e checks if two values are equal.
- "is" operator known as identity operator, checks if two variables point to the exact same object in memory.

**8) What are logical operators in Python?**
- Logical operators helps in improving decision making scenarios.There are 3 logical operators in python (and, or, not)
- and operator- returns true if both the conditions are true.
- or operator- return true if atleast one condtion is true.
- not opearator- returns true if condtion is false and vice-versa.


**9) What is type casting in Python?**
- Type casting in python means to convert one data type into another. There are two types of type casting in python- implicit and explicit. Implicit typecasting is done automatically by python while explicit type casting is done by the programmer using in-built functions like str,int,float and bool.

**10) What is the difference between implicit and explicit type casting?**
- **Implicit typecasting** - is done automatically by the compiler/interpreter. It does not require manual intervention from the programmer. Usually occurs when a value of a smaller data type is converted to a larger data type to prevent data loss.
- **Explicit typecasting**- Requires manual intervention by the programmer.
Python converts the datatype using inbuilt functions such as str, int, float and bool.

**11) What is the purpose of conditional statements in Python?**
- The purpose of conditional statements is that they help in making decisions, thus executing code based on the condtions given by users.
- These statements help in creating interactive and dynamic programs.
- if, elif, and else are used to write these conditional statements.

**12) How does the elif statement work?**
- 'elif' statement stands for else-if.
- It is used to check multiple conditions after the initial if and before the final else.
- It's used when we want to test several possible conditions and run different blocks of code depending on which condition is true, thus helps to avoid writing many if statements.

**13) What is the difference between for and while loops?**
- 'for' loop is used when the number of iterations is known whereas 'while' loop is used where condition based repetition is needed.
- 'for' works with sequences like list, string, range. 'while' runs until a condition is false.

**14) Describe a scenario where a while loop is more suitable than a for loop.**
- A while loop is more suitable than a for loop where the number of iterations is not known in advance.
- Suppose we have made an app that requires you to sign in using your password that says the password should contain atleast 8 letters. But we donot know that user will enter the correct password after how many attempts, then while loop will be a better choice since we don not know after how many iterations, the program should end.
